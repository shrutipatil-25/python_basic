For example print first n (ask user Input) even numbers using the while loop in Python


#example



Enter a number of even numbers you want to see from start: 5

2, 4, 6,8,10

```python
numbers = int(input('Enter a number from which you want to check even numbers'))
while numbers <=10:
  if numbers % 2 == 0:
    print(numbers)
  numbers += 1
```

Python program to print Even Numbers in given range

```python
start = int(input("Enter the start of range: "))
end = int(input("Enter the end of range: "))

for num in range(start,end+1):
  if num % 2 == 0:
    print(num)
```
